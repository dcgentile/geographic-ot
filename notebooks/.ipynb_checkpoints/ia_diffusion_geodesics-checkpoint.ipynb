```python
import total
import os
import numpy as np
import pandas as pd 
import seaborn as sns
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import geopandas
from libpysal import weights
from gerrychain import Graph

img_dir = os.path.join(".", "geodesic_stills")
```

```python
filename = "./IA_counties/IA_counties.shp"
voter_df = pd.read_csv("./data/voter_data.csv")
ia_graph = Graph.from_file(filename)
ia_counties = geopandas.read_file(filename)
ia_adj=nx.adjacency_matrix(ia_graph).toarray()
centroids = np.column_stack((ia_counties.centroid.x, ia_counties.centroid.y))
queen = weights.Queen.from_dataframe(ia_counties)
graph = queen.to_networkx()
positions = dict(zip(graph.nodes, centroids))
num_nodes = len(graph.nodes)
```

```python
def compute_graph_resistance(graph):
    ia_laplace = nx.laplacian_matrix(ia_graph).toarray()
    mpinv = np.linalg.pinv(ia_laplace)
    num_nodes = len(graph.nodes)
    metric = np.zeros((num_nodes, num_nodes))
    for i in range(num_nodes):
        for j in range(num_nodes):
            delta = np.zeros(num_nodes)
            delta[i] = 1
            delta[j] = -1
            metric[i,j] = (delta.T @ mpinv @ delta) 
    return metric

def compute_graph_metric(graph):
    graph_dists = dict(nx.shortest_path_length(graph))
    sorted_dists = {k:sorted(v.items()) for k, v in graph_dists.items()}
    num_nodes = len(graph.nodes)
    dist_mat = np.zeros((num_nodes, num_nodes))
    for i in range(num_nodes):
        for j in range(num_nodes):
            dist_mat[i,j] = sorted_dists[i][j][1]
    
    return dist_mat

def iter_experiment(its_lo, its_hi, its_step, measures, cost, epsilon, steps):
    for its in range(its_lo, its_hi, its_step):
        tot_geodesic = total.build_geodesic(measures, cost, epsilon, loss="sqeuc", method="total", steps=steps, iters=its)
        pot_geodesic = total.build_geodesic(measures, cost, epsilon, method="", steps=steps)
        #tot_ratio = np.divide(tot_geodesic, pot_geodesic, out=np.zeros_like(pot_geodesic))
        tot_ratio = tot_geodesic/pot_geodesic
        tot_rel_err = abs(np.ones((num_nodes, steps)) - tot_ratio)
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
        fig.set_size_inches(16,8)
        sns.heatmap(tot_geodesic, ax=ax1)
        sns.heatmap(pot_geodesic, ax=ax2)
        sns.heatmap(tot_rel_err, ax = ax3)
        return fig, tot_rel_err
```

```python
def comparison_video(measures, cost, epsilon, steps, its, filename, metric, p=2):
    w1_geodesic = total.build_geodesic(measures, cost, epsilon, method="total", steps=steps, iters=its)
    w2_geodesic = total.build_geodesic(measures, cost ** p, epsilon, method="total", steps=steps, iters=its)
    pot_w1_geodesic = total.build_geodesic(measures, cost, epsilon, method="", steps=steps, iters=its)
    pot_w2_geodesic = total.build_geodesic(measures, cost ** p, epsilon, method="", steps=steps, iters=its)
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
    fig.set_size_inches(20,20)
     
    dir_name = "p-comparison-anims/"
    still_dir = os.path.join(img_dir, dir_name)
    if not os.path.isdir(still_dir):
        os.mkdir(still_dir)
     
    def update(frame):
        size = 500
        nx.draw(graph, positions, ax=ax1, node_size=size, node_color=w2_geodesic[:, frame])
        nx.draw(graph, positions, ax=ax2, node_size=size, node_color=w1_geodesic[:, frame])
        nx.draw(graph, positions, ax=ax3, node_size=size, node_color=pot_w2_geodesic[:, frame])
        nx.draw(graph, positions, ax=ax4, node_size=size, node_color=pot_w1_geodesic[:, frame])
        ax1.set_title(
            f"W{p}: {metric}\n Timestep : {frame + 1}"
            )
        ax2.set_title(
            f"W1: {metric}\n Timestep : {frame + 1}"
            )
        ax3.set_title(
            f"POT W{p}: {metric}\n Timestep : {frame + 1}"
            )
        ax4.set_title(
            f"POT W1: {metric}\n Timestep : {frame + 1}"
            )
    ani = animation.FuncAnimation(fig=fig, func=update, frames=steps, interval=100)
    ani.save(os.path.join(dir_name, filename), writer='ffmpeg')

def draw_endpoints(measures, cost, epsilon, steps, its, filename, metric):
    w1_geodesic = total.build_geodesic(measures, cost, epsilon, method="total", steps=steps, iters=its)
    w2_geodesic = total.build_geodesic(measures, cost ** 2, epsilon, method="total", steps=steps, iters=its)
    pot_w1_geodesic = total.build_geodesic(measures, cost, epsilon, method="", steps=steps, iters=its)
    pot_w2_geodesic = total.build_geodesic(measures, cost ** 2, epsilon, method="", steps=steps, iters=its)
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
    fig.set_size_inches(20,20)
     
    dir_name = "p-comparison-stills/"
    still_dir = os.path.join(img_dir, dir_name)
    if not os.path.isdir(still_dir):
        os.mkdir(still_dir)
    
    lo, hi = 0, steps - 1
    size=500
    nx.draw(graph, positions, ax=ax1, node_size=size, node_color=w2_geodesic[:, lo])
    nx.draw(graph, positions, ax=ax2, node_size=size, node_color=w1_geodesic[:, hi])
    nx.draw(graph, positions, ax=ax3, node_size=size, node_color=w2_geodesic[:, lo])
    nx.draw(graph, positions, ax=ax4, node_size=size, node_color=w1_geodesic[:, hi])
    ax1.set_title(
        f"W2: {metric}\n Timestep : {lo + 1}"
        )
    ax2.set_title(
        f"W1: {metric}\n Timestep : {hi + 1}"
        )
    ax3.set_title(
        f"W2: {metric}\n Timestep : {lo + 1}"
        )
    ax4.set_title(
        f"W1: {metric}\n Timestep : {hi + 1}"
        )
    plt.savefig(os.path.join(dir_name, filename))
```
